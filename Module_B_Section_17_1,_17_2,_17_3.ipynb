# 17.1: Interpolation Problem Statement

Example 1: Plot the points (0,2), (4,4), (5,1), and (7,6).

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [0, 4, 5, 7]
y = [2, 4, 1, 6]



plt.figure(figsize = (10,8))
plt.plot(x, y, 'ob') #makes the points blue
plt.title('Linear Interpolation at x = 3.5') #names the plot 
plt.xlabel('x') #lables the x axis 
plt.ylabel('y') #lables the y axis
plt.show()
```

# 17.2: Linear Interpolation 

Example 2: Using linear interpolation, plot the points (0,2), (4,4), (5,1), and (7,6) and find the corresponding y-hat at x=3.5.

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [0, 4, 5, 7]
y = [2, 4, 1, 6]

f = interp1d(x, y)
y_hat = f(3.5) #the corresponding y value at x=3.5 on the new y_hat function
print(y_hat) #prints 2.833333333333333

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob') #adds the lines and makes both the points and lines blue
plt.plot(3.5, y_hat, 'ro') #adds the coordinate (3.5,y_hat) to plot in red 
plt.title('Linear Interpolation at x = 3.5') #names the plot 
plt.xlabel('x') #lables the x axis 
plt.ylabel('y') #lables the y axis
plt.show()
```

# 17.3: Cubic Spline Interpolation

Example 3: Using cubic spline interpolation, plot the points (0,2), (4,4), (5,1), and (7,6) and find the corresponding point on the new interpolation equation at x=3.5.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = [0, 4, 5, 7]
y = [2, 4, 1, 6]

# use bc_type = 'natural' adds the constraints as we described above
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(-1, 8, 100)  # shows range from -1<x<8
y_new = f(x_new)
y_hat = f(3.5) #the corresponding y value at x=3.5 on the new y_hat function
print(y_hat) #prints 4.8984375

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b') #makes the line blue
plt.plot(x, y, 'bo') #makes the points blue
plt.plot(3.5, y_hat, 'ro') #adds the coordinate (3.5,y_hat) to plot in red
plt.title('Cubic Spline Interpolation at x=3.5')
plt.xlabel('x') #lables the x axis 
plt.ylabel('y') #lables the y axis 
plt.show()
```
